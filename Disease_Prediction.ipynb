## Step - 1: Data Loading

To begin the process of building the machine learning model for disease prediction with MediAI, the first step is to load the dataset into the notebook environment. The dataset, named Data.csv, contains information about various symptoms and the corresponding prognosis (0 or 1) for each case.

We will use pandas, a powerful data manipulation library in Python, to load the dataset into a DataFrame. The DataFrame will allow us to easily explore and manipulate the data as we proceed with our analysis and model building.


```python
dataset = pd.read_csv('data.csv')
```

```python
# 2.6 Distribution of Prognosis in the Dataset
plt.figure(figsize=(20, 20))
prognosis_counts = dataset['prognosis'].value_counts()
prognosis_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='Paired')
plt.title('Prognosis Distribution')
plt.ylabel('')
```

```python
# 2.7 Visualizing the symptoms of different diseases 
symptoms_vs_prognosis = dataset.drop('prognosis', axis=1).groupby(dataset['prognosis']).sum().T
# Iterate over each prognosis
for prognosis_to_visualize in symptoms_vs_prognosis.columns:
    # Filter the symptoms for the selected prognosis
    symptoms_for_prognosis = symptoms_vs_prognosis[prognosis_to_visualize].loc[symptoms_vs_prognosis[prognosis_to_visualize] > 0]

    # Create subplots
    fig, ax = plt.subplots(figsize=(8, 5))
    
    # Plotting the symptoms for the selected prognosis with different colors for bars
    colors = plt.cm.tab10(range(len(symptoms_for_prognosis)))
    symptoms_for_prognosis.plot(kind='barh', color=colors, ax=ax)
    ax.set_title(f'Symptoms for {prognosis_to_visualize}', fontsize=16)
    ax.set_ylabel('Symptoms', fontsize=14)
    ax.set_xlabel('Count', fontsize=14)
    ax.tick_params(axis='y', rotation=0, labelsize=12)
    ax.tick_params(axis='x', labelsize=12)
    ax.grid(axis='x', linestyle='--', alpha=0.7)
    
    # Save the individual figure
    plt.savefig(f'Data_Insights/Symptoms_for_{prognosis_to_visualize}.png', bbox_inches='tight')
    
    # Show the figure
    plt.show()
```

## Step 3: Splitting the Dataset into Training and Test Data
Before training our machine learning models for disease prediction with MediAI, it is essential to split the dataset into training and test sets. This allows us to train the models on a subset of the data and evaluate their performance on unseen data to assess generalization.
### Key Considerations:
1. **Training Set**: The training set comprises a subset of the dataset used to train the machine learning models. It typically accounts for a significant portion of the data (e.g., 70-80%).
2. **Test Set**: The test set is used to evaluate the performance of the trained models on unseen data. It helps assess how well the models generalize to new observations.
3. **Randomization**: It is crucial to randomly shuffle the data before splitting to ensure that the training and test sets are representative of the overall dataset.
4. **Stratification**: In classification tasks with imbalanced classes, it is advisable to use stratified sampling to ensure that the class distribution is preserved in both the training and test sets.


```python
from sklearn.model_selection import train_test_split
X = dataset.drop(columns=['prognosis'])  # Features
y = dataset['prognosis']  # Target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Step 4: Building the ML Model

With the dataset loaded, explored, and split into training and test sets, we can proceed to build machine learning models for disease prediction with MediAI. In this step, we will select appropriate algorithms, train the models using the training data, and evaluate their performance using the test data.

### Key Steps in Building the ML Model:

1. **Selecting Algorithms**: Choose suitable machine learning algorithms based on the nature of the problem (classification in this case), the dataset size, and the desired interpretability of the models.

2. **Training the Models**: Fit the selected algorithms to the training data using the `fit` method. This involves learning the patterns and relationships in the data.

3. **Model Evaluation**: Evaluate the performance of the trained models using appropriate evaluation metrics such as accuracy, precision, recall, F1-score, and area under the ROC curve (AUC-ROC).

### 4.1 Logistic Regression Model Training

```python
# Model Building and Fitting the Model 
from sklearn.linear_model import LogisticRegression
# Initialize the Logistic Regression model
logistic_regression = LogisticRegression()
# Fit the model to the training data
logistic_regression.fit(X_train, y_train)
```

```python
# Checking for Overfitting and UnderFitting 
# Accuracy on the training set
train_accuracy = logistic_regression.score(X_train, y_train)
print("Training Accuracy:", train_accuracy)
# Accuracy on the test set
test_accuracy = logistic_regression.score(X_test, y_test)
print("Test Accuracy:", test_accuracy)
```

```python
# Making Predictions of X_test 
y_pred = logistic_regression.predict(X_test)
```

```python
# Accuracy of Model
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
# Making predictions on the test set
y_pred_lc = logistic_regression.predict(X_test)
# Calculate accuracy
accuracy_lc = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy_lc)
```

```python
# Saving the Model
from joblib import dump

# Save the model to a file
model_file = r'Model/logistic_regression_model.joblib'
dump(logistic_regression, model_file)
print("Model saved to:", model_file)
```

### Reasons Not to Choose Logistic Regression 
1. **Target is Multiclass**: The classification problem involves predicting multiple classes, meaning there are more than two possible outcomes that the model can predict.
2. **Average='binary'**: This parameter specifies the method of aggregation for metrics in multiclass settings. However, 'binary' average is suitable only for binary classification problems where there are two classes (e.g., 0 and 1).
3. **Error Explanation**: The error occurs because 'binary' average is incompatible with multiclass classification. It attempts to aggregate metrics across different classes as if they were binary outcomes, leading to the ValueError.
